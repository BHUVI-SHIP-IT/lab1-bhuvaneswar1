```python
# prompt: Lab Question 1: Performance Analysis of Linear Search on Best, Average, and Worst Cases
# Problem Statement:
# Write a program to implement the linear search algorithm. Analyze and compare the time taken for:
#  Best case (element at the first position),
#  Average case (element in the middle),
#  Worst case (element not present).
# 
# Instructions:
#  Generate a list of n random integers (e.g., n = 10000).
#  Measure the time taken for searching the key in each of the three cases.
#  Plot a graph comparing the time taken vs. the case type.
# Expected Analysis Output:
#  Explain why the best case is O(1), average case is O(n/2), and worst case is O(n).
#  Include actual execution times in milliseconds.
#  Provide observations on the impact of input size n.
# provide user to input array size and input search key

import time
import random
import matplotlib.pyplot as plt

def linear_search(arr, key):
  """Implements the linear search algorithm."""
  for i in range(len(arr)):
    if arr[i] == key:
      return i  # Return the index if found
  return -1 # Return -1 if not found

# Get input from the user
array_size = int(input("Enter the size of the array (n): "))
search_key = int(input("Enter the key to search for: "))

# Generate a list of random integers
data = [random.randint(1, array_size * 2) for _ in range(array_size)]

# --- Performance Analysis ---

# Best Case: Element at the first position
best_case_data = data[:] # Create a copy to avoid modifying the original
best_case_data[0] = search_key
start_time = time.time()
linear_search(best_case_data, search_key)
best_case_time = (time.time() - start_time) * 1000 # Time in milliseconds

# Average Case: Element in the middle
# Ensure the key exists in the middle
average_case_data = data[:]
average_index = array_size // 2
average_case_data[average_index] = search_key
start_time = time.time()
linear_search(average_case_data, search_key)
average_case_time = (time.time() - start_time) * 1000 # Time in milliseconds


# Worst Case: Element not present (or at the end, for simplicity we use not present)
worst_case_data = data[:] # Create a copy
# Ensure the key is not in the worst_case_data
while search_key in worst_case_data:
    search_key = random.randint(1, array_size * 2) # Generate a new key not in the list
start_time = time.time()
linear_search(worst_case_data, search_key)
worst_case_time = (time.time() - start_time) * 1000 # Time in milliseconds

# --- Output and Analysis ---

print("\n--- Performance Analysis of Linear Search ---")
print(f"Array Size (n): {array_size}")
print(f"Search Key: {search_key}")
print(f"Best Case Time (Element at first position): {best_case_time:.4f} ms")
print(f"Average Case Time (Element in the middle): {average_case_time:.4f} ms")
print(f"Worst Case Time (Element not present): {worst_case_time:.4f} ms")

print("\n--- Expected Analysis ---")
print("Best Case (O(1)): The element is found at the very beginning of the list. The loop runs only once.")
print("Average Case (O(n/2)): On average, the element is found somewhere in the middle of the list. The loop runs approximately half the length of the list.")
print("Worst Case (O(n)): The element is either at the very end of the list or not present at all. The loop has to iterate through the entire list.")

# Plotting the results
cases = ['Best Case', 'Average Case', 'Worst Case']
times = [best_case_time, average_case_time, worst_case_time]

plt.figure(figsize=(8, 6))
plt.bar(cases, times, color=['green', 'orange', 'red'])
plt.ylabel('Time (milliseconds)')
plt.title(f'Linear Search Performance for n = {array_size}')
plt.show()

print("\n--- Observations ---")
print(f"Impact of input size n={array_size}: As the input size (n) increases, the difference in time between the best case and the worst case becomes more significant. The linear relationship between the number of operations and n in the average and worst cases is evident in the measured times.")
```
